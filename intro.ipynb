Example 1: Find $\sqrt{2}$

Analysis: equivalent task is to solve the equation $x^2 - 2 = 0$ <br>

Clearly $x > 1$, so lets write $x = 1 + \delta$ with the idea that $\delta$ is small <br>
Then: $1 + 2\delta + \delta^2 = 2$ <br>

Solving this equation still involves a square root<br>
Find an approximate solution by ignoring $\delta^2$ term: then $\delta = 1/2$ <br>
New value of $x = 1 + \delta = 3/2$ <br>

We can improve the approximation by using the new value of $x$ as the starting guess: $x = 3/2 + \delta$<br>


Algorithm: Steps to solution

$_0=1$  &emsp; &emsp; &emsp; &emsp; &emsp; # Initial value

$_0^2+2_0=2$  &emsp; &emsp; &emsp; &emsp; &emsp; # Linearized equation in $\delta$ (approximate)

$=1/_0_0/2$  &emsp; &emsp; &emsp; &emsp; &emsp;  # Calculate $\delta$

$_1=_0+=1/_0+_0/2=1.5$  &emsp; &emsp; &emsp; &emsp; &emsp;   &emsp; &emsp;    # Improved x

$_2=1/_1+_1/2=1.4167$    &emsp; &emsp; &emsp; &emsp; &emsp;  &emsp; &emsp; &emsp; # Repeat


Repeating the steps leads to a smaller error - process called iteration

1) Useful to automate the repetition of operations $\rightarrow$ <u>assignments</u>, <u>loops</u>

2) Two sources of error in numerical approximations:<br>
&emsp; &emsp; a) Round off - how many digits we calculate<br>
&emsp; &emsp; b) Algorithmic - here the effect of ignoring the $\delta^2$ terms<br>
&emsp; Usually algorithm error is most important - check for convergence $\rightarrow$ <u>conditionals</u>
